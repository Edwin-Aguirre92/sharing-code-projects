#### Cramer's rule 

```python
###create a function of an n by n matrix to obtain the determinant for 2 or 3 size square matrix

def det_1(m):
    ##obtaining the determinant of a matrix, if the number of entries within a matrix is ==4 
    if m.size ==4:
        m1= m[0,0]*m[1,1]-m[1,0]*m[0,1]
    ##if the number of entries within a matrix is not equal to 4:
    else:
        m1=m[0,0]*(m[1,1]*m[2,2]-m[2,1]*m[1,2])-m[0,1]*(m[1,0]*m[2,2]-m[2,0]*m[1,2])+m[0,2]*(m[1,0]*m[2,1]-m[2,0]*m[1,1])
    return m1
```

```python
#function that uses the cram rule methodology and outputs the solution to the system of equations.
#It takes as input the original matrix, the column vector, and the determinant function. 
def cram_rule(mt,col1,det_1):
    #creating copies of original matrix
    mx1=mt.copy()
    mx2=mt.copy()
    mx3=mt.copy()
    for i in range(len(mx1)):
        mx1[i,0]=col1[i]
        mx2[i,1]=col1[i]
        mx3[i,2]=col1[i]
    #obtaining the determinant of all matrices
    D1=(det_1(mx1)/det_1(mt))
    D2=(det_1(mx2)/det_1(mt))
    D3=(det_1(mx3)/det_1(mt))
    return D1,D2,D3
```

### Cramer's Rule( basically, states that for each unknown of a system one can express the solution of an unknown as a fraction of the determinant of the system)

```python
#read image 
img = mpimg.imread('LinearSystem.png')
#resizing the image 
plt.figure(figsize=(10,10))
#showing the image
plt.imshow(img) 
```

```python
#1-create matrix for system of linear equations
mt=np.zeros((3,3))
print(mt)
mt[0,0]=0.3
mt[0,1]=0.52
mt[0,2]=1
mt[1,0]=0.5
mt[1,1]=1
mt[1,2]=1.9
mt[2,0]=0.1
mt[2,1]=0.3
mt[2,2]=0.5
print(mt)
```

```python
#2-getting determinant of said system
D=det_1(mt)
print(D)
```

```python
#3- obtaining unknown x1,x2,x3 as a function of the determinant, by replacing x1 with the column vector given in the problem.
col_z=np.zeros(3)
col_z[0]=-0.01
col_z[1]=0.67
col_z[2]=-0.44
#using the previously defined rule. 
f5=cram_rule(mt,col_z,det_1)
print("The unknowns for the previously defined linear system of equation  are:",round(f5[0],2),round(f5[1],2),round(f5[2],2))
```

### Solving a problem of linear equations using cramer's rule

```python
#constants for problem
g=9.80
v=5
c1=10
c2=14
c3=17
m1=70
m2=60
m3=40
```

```python
#Problem matrix
mp=np.zeros((3,3))
```

```python
mp[0,0]=70
mp[1,0]=60
mp[2,0]=40
mp[0,1]=1
mp[1,1]=-1
mp[1,2]=1
mp[2,2]=-1
print(mp)
```

```python
b_p=np.zeros(3)
b_p[0]=m1*g-c1*v
b_p[1]=m2*g-c2*v
b_p[2]=m3*g-c3*v
print(b_p)
```

```python
##Using Cramer's rule 
f3=cram_rule(mp,b_p,det_1)
print("The unknowns for the previously defined linear system of equation  are:",round(f3[0],2),round(f3[1],2),round(f3[2],2))
```
